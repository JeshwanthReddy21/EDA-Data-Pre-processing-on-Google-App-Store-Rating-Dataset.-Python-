### 6. Check if there are missing values present in the column Rating, If any? drop them and and create a new column as 'Rating_category' by converting ratings to high and low categories(>3.5 is high rest low)

```python
apps_dataset['Rating'].isnull().sum()#There are 1465 missing values
```

```python
# Droping the missing values in Rating column
apps_dataset.dropna(subset=['Rating'], inplace=True)
```

```python
apps_dataset.shape
```

```python
# Creating a new column as 'Rating_category' by converting ratings to high and low categories
def rating_category(rating):
    if rating > 3.5:
        return 'High'
    else:
        return 'Low'
    
apps_dataset['Rating_category'] = apps_dataset['Rating'].apply(rating_category)
```

```python
apps_dataset.shape
```

```python
apps_dataset
```

```python
# Get the number of low in Rating_category column
low_count = apps_dataset['Rating_category'].value_counts()['Low']

# Print the result
print(low_count)
```

### 9. The column 'Size' contains alphanumeric values, treat the non numeric data and convert the column into suitable data type. (hint: Replace M with 1 million and K with 1 thousand, and drop the entries where size='Varies with device')

```python
apps_dataset['Size'] = apps_dataset['Size'].str.replace('M', '1000000')
apps_dataset['Size'] = apps_dataset['Size'].str.replace('k', '1000')
```

```python
apps_dataset = apps_dataset[apps_dataset['Size'] != 'Varies with device']
```

### 10. Check the column 'Installs', treat the unwanted characters and convert the column into a suitable data type.

```python
apps_dataset["Installs"] #There are string type in Install column so we are removing the string in Install
```
